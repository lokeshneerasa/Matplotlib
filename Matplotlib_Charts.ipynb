```python
########################### MATPLOTLIB #########################################
######################### 1.Line Chart ########################################
# Step-1.Create Simple LineChart
import matplotlib.pyplot as plt
x=[7,8]
y=[7,8]
plt.plot(x,y)
plt.show()
```

```python
# Step-2.Creating Simple line Chart using marker and markersize
import matplotlib.pyplot as plt
x=[7,8]
y=[7,8]
plt.plot(x,y,marker='h',markersize=10)
plt.show()
```

```python
# Step-3.Creating Simple line Chart with different values and give title
import matplotlib.pyplot as plt
x=[5,9,20]
y=[5,9,12]
plt.plot(x,y,marker='*')
plt.title("A Simple Line Chart")
plt.show()
```

```python
# Step-4 Give labels to linegraph(xlabel,ylabel)
import matplotlib.pyplot as plt
x=[7,8,10]
y=[7,8,20]
plt.plot(x,y,marker='s',markersize=10)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Step-5 Create two lines in single chart
import matplotlib.pyplot as plt
x=[7,8,10]
y=[7,8,20]
p=[10,14,18]
plt.plot(x,y,marker='o')
plt.plot(x,p,marker='+')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Double Lines")
plt.show()
```

```python
# Step-6 Create Multiple Linecharts in Single chart
import matplotlib.pyplot as plt
x=["mon","Tue","wed"]
y=[9,12,15]
p=[10,12,16]
q=[9,13,14]
plt.plot(x,y,marker='+',markersize=10)
plt.plot(x,p,marker='v',markersize=10)
plt.plot(x,q,marker='x',markersize=10)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Multiple Lines")
plt.grid(True)
plt.show()
```
